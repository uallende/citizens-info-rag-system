```python
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pdfkit

# Define Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")

user_home_dir = os.path.expanduser("~")
chrome_binary_path = os.path.join(user_home_dir, "chrome-linux64", "chrome")
chromedriver_path = os.path.join(user_home_dir, "chromedriver-linux64", "chromedriver")

chrome_options.binary_location = chrome_binary_path
service = Service(chromedriver_path)

url = 'https://www.citizensinformation.ie/en/health/legal-matters-and-health/assisted-decision-making-act/'

with webdriver.Chrome(service=service, options=chrome_options) as driver:
    driver.get(url)

    driver.execute_script("document.cookie = 'cookiecheck=true';")
    cookie = driver.get_cookie('cookiecheck')
    print(cookie)  # Should print {'name': 'cookiecheck', 'value': 'true', ...}
    WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.TAG_NAME, 'body')))

    html = driver.page_source

config = pdfkit.configuration(wkhtmltopdf='/usr/bin/wkhtmltopdf')
output_file = 'out.pdf'

# pdfkit.from_string(html, output_file, configuration=config, verbose=True)
pdfkit.from_string(html, output_file, configuration=config)
```

```python
from langchain_community.document_loaders import WebBaseLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter, CharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
import os
import pdfkit

# pdfkit.from_url("https://www.citizensinformation.ie/en/health/legal-matters-and-health/assisted-decision-making-act/", "output.pdf")
current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)
path_to_pdf = 'notebooks'
file_path = os.path.join(parent_dir, path_to_pdf)

documents_text = []
c = 0

for doc in os.listdir(file_path):
    # if file ends in .pdf
    if not doc.endswith('.pdf'):
        continue
    doc_path = os.path.join(file_path, doc)
    loader = PyPDFLoader(doc_path)
    pages = loader.load_and_split()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 0)
    docs = text_splitter.split_documents(pages)
    documents_text.append(docs)
    c += 1
    if c == 5:
        break

documents_text = [item for sublist in documents_text for item in sublist]
text = ""

for doc in documents_text:
    text += doc.page_content + "\n"

print(text)
```
